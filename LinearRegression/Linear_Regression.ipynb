```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
import numpy as np
```

```python
x = np.array(x)
y = np.array(y)
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
lr_model.fit(x,y)
```

```python
v_predi = lr_model.predict(x)
v_predi
```

```python
y_3 = lr_model.predict([[3.5]])
y_3
```

```python
coef = lr_model.coef_
intercept = lr_model.intercept_
print(coef,intercept)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
MSE = mean_squared_error(y,v_predi)
R2 = r2_score(y,v_predi)
print(MSE,R2)
```
